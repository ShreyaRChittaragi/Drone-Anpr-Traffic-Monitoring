```python
print("üî§ Initializing EasyOCR reader...")
reader = easyocr.Reader(['en'], gpu=True)
print("‚úÖ EasyOCR reader initialized!")
```

```python
print("\nüé¨ Starting video processing...")
print("=" * 60)

# Open video file
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"üìπ Video FPS: {fps}")
print(f"üìπ Total frames: {total_frames}")

# Process every Nth frame to avoid overwhelming output
frame_skip = 30  # Process every 30th frame (1 per second for 30fps video)
frame_count = 0
detected_plates = []

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    frame_count += 1

    # Skip frames
    if frame_count % frame_skip != 0:
        continue

    print(f"\nüìç Processing frame {frame_count}/{total_frames}...")

    # Run YOLO detection
    results = model(frame, conf=0.5)

    # Process detections
    for result in results:
        boxes = result.boxes

        for box in boxes:
            # Get bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = float(box.conf[0])

            # Crop license plate region
            plate_img = frame[y1:y2, x1:x2]

            if plate_img.size == 0:
                continue

            # Use EasyOCR to extract text
            ocr_results = reader.readtext(plate_img)

            plate_text = ""
            if ocr_results:
                plate_text = " ".join([text[1] for text in ocr_results])
                print(f"   üöó DETECTED PLATE: {plate_text} (Confidence: {confidence:.2f})")
                detected_plates.append({
                    'frame': frame_count,
                    'plate': plate_text,
                    'confidence': confidence,
                    'timestamp': datetime.utcnow().isoformat()
                })

            # Draw bounding box and text on frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, plate_text, (x1, y1-10),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Display frame using matplotlib
    if frame_count % (frame_skip * 2) == 0:  # Show every 2nd processed frame
        plt.figure(figsize=(12, 8))
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.title(f"Frame {frame_count} - License Plate Detection")
        plt.axis('off')
        plt.show()

cap.release()

print("\n" + "=" * 60)
print(f"‚úÖ Video processing complete!")
print(f"üìä Total plates detected: {len(detected_plates)}")

# Display all detected plates
if detected_plates:
    print("\nüìã DETECTED LICENSE PLATES SUMMARY:")
    print("-" * 60)
    for i, plate in enumerate(detected_plates, 1):
        print(f"{i}. Frame {plate['frame']}: {plate['plate']} "
              f"(Conf: {plate['confidence']:.2f})")
```

```python
print("\n" + "=" * 60)
print("üíæ MOCK DYNAMODB WRITE DEMONSTRATION")
print("=" * 60)

def save_to_dynamodb(plate_data):
    """
    Demonstrates how detected plates would be saved to DynamoDB
    in the production environment.
    """
    current_time = datetime.utcnow()

    print(f"\nüìù Saving plate to DynamoDB:")
    print(f"   Plate Number: {plate_data['plate']}")
    print(f"   Timestamp: {current_time.isoformat()}")
    print(f"   Frame: {plate_data['frame']}")
    print(f"   Confidence: {plate_data['confidence']:.2f}")

    print("""
    üîµ ACTUAL AWS DYNAMODB CODE:

    import boto3
    from datetime import datetime

    dynamodb = boto3.resource('dynamodb', region_name='us-east-1')
    table = dynamodb.Table('DetectedPlates')

    def save_plate(plate_number, confidence, frame_id):
        response = table.put_item(
            Item={
                'PlateID': f"{plate_number}_{int(time.time())}",
                'PlateNumber': plate_number,
                'Timestamp': datetime.utcnow().isoformat(),
                'Confidence': str(confidence),
                'FrameID': frame_id,
                'StreamSource': 'drone-01',
                'Location': 'geo-coordinates-here',
                'ProcessedBy': 'ANPR-Processor-v1'
            }
        )
        return response
    """)

    print("‚úÖ Mock save complete!")

# Demonstrate saving detected plates
if detected_plates:
    print("\nüìä Saving all detected plates to DynamoDB (MOCK):")
    for plate in detected_plates[:3]:  # Save first 3 as examples
        save_to_dynamodb(plate)
        time.sleep(0.5)
```
